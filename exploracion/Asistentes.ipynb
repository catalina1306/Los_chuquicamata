```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_asistentes_ = pd.read_csv("https://drive.google.com/uc?id=1btHx48Rz2-Eewk3lrZKXRdg2AobmA2oe", sep=None, engine='python')
```

```python
df_asistentes = df_asistentes_[(df_asistentes_['COD_REG_RBD']==13) & 
                (df_asistentes_['ESTADO_ESTAB']==1) &
                (df_asistentes_['N_ASIS']>0)
                ]
df_asistentes.columns = df_asistentes.columns.str.replace('\ufeff', '', regex=False)
df_asistentes.drop(columns=['ESTADO_ESTAB','NOM_REG_RBD_A', 'AGNO','COD_DEPROV_RBD', 'NOM_DEPROV_RBD','DGV_RBD','COD_REG_RBD','COD_PRO_RBD','COD_DEPE2','COD_DEPE','COD_COM_RBD'], inplace=True)
df_asistentes.head()
```
