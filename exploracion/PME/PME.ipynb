Bastantes resultados para hacer analisis especifico pero encontramos resultados interesantes de correlacion en cuanto a ESTIM_TOTAL y fijamos un interes en

```python
def detectar_outliers_iqr(data, columnas):
    outliers = {}

    for columna in columnas:
        Q1 = data[columna].quantile(0.25)
        Q3 = data[columna].quantile(0.75)
        IQR = Q3 - Q1

        lim_inferior = Q1 - 1.5 * IQR
        lim_superior = Q3 + 1.5 * IQR

        outliers_columna = data[(data[columna] < lim_inferior) | (data[columna] > lim_superior)]
        outliers[columna] = outliers_columna

    return outliers
# Detectar outliers
outliers = detectar_outliers_iqr(df, df_numeric.columns)

# Mostrar el número de outliers por columna
for columna, datos_outliers in outliers.items():
    print(f'Columna: {columna}, Número de outliers: {datos_outliers.shape[0]}')
```
